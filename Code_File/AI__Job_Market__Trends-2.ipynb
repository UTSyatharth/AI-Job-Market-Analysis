```python
# =====================================================
# 🧠 AI/ML Job Market Data Analysis & Visualization
# =====================================================
# Author: Yatharth Sharma
# Description: This notebook explores trends, salaries,
# skills, and hiring patterns in the AI/ML job market.
# =====================================================

# -----------------------------------------------------
# 🔧 Step 1. Setup & Import Libraries
# -----------------------------------------------------
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from wordcloud import WordCloud
from pyspark.sql import SparkSession
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import plotly.express as px

# Mount Google Drive to access files
drive.mount('/content/drive')

# -----------------------------------------------------
# 📂 Step 2. Load the Dataset
# -----------------------------------------------------
df = pd.read_csv('/content/drive/MyDrive/Python/ai_job_market.csv')
display(df.head())

# -----------------------------------------------------
# 🧭 Step 3. Explore the Dataset
# -----------------------------------------------------
print("Dataset shape:", df.shape)
print("\nColumns:", list(df.columns))
print("\nFirst 5 rows:")
display(df.head())

# -----------------------------------------------------
# 🧹 Step 4. Data Cleaning
# -----------------------------------------------------
# Convert 'posted_date' to datetime
df['posted_date'] = pd.to_datetime(df['posted_date'], errors='coerce')

# Function to extract median salary from salary range
def extract_salary(s):
    if pd.isna(s):
        return np.nan
    numbers = re.findall(r'\d+', str(s).replace(',', ''))
    if len(numbers) == 2:
        return (int(numbers[0]) + int(numbers[1])) / 2
    elif len(numbers) == 1:
        return int(numbers[0])
    else:
        return np.nan

df['salary_median'] = df['salary_range_usd'].apply(extract_salary)

# -----------------------------------------------------
# 📊 Step 5. Basic Explorations
# -----------------------------------------------------
# Top 10 most common job titles
top_jobs = df['job_title'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette="viridis")
plt.title('Top 10 Job Titles')
plt.xlabel('Number of Postings')
plt.ylabel('Job Title')
plt.show()

# Experience level distribution
plt.figure(figsize=(8,4))
sns.countplot(y='experience_level', data=df,
              order=df['experience_level'].value_counts().index,
              palette='coolwarm')
plt.title('Experience Level Distribution')
plt.xlabel('Count')
plt.ylabel('Experience Level')
plt.show()

# Top 10 job locations
top_locations = df['location'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.values, y=top_locations.index, palette="cubehelix")
plt.title('Top 10 Locations')
plt.xlabel('Number of Postings')
plt.ylabel('Location')
plt.show()

# -----------------------------------------------------
# 💰 Step 6. Salary Analysis
# -----------------------------------------------------
# Salary distribution
plt.figure(figsize=(8,4))
sns.histplot(df['salary_median'].dropna(), bins=30, kde=True, color='teal')
plt.title('Salary Distribution')
plt.xlabel('Salary (USD)')
plt.ylabel('Count')
plt.show()

# Median salary by top job titles
top_salary_titles = df.groupby('job_title')['salary_median'].median().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_salary_titles.values, y=top_salary_titles.index, palette="magma")
plt.title('Top 10 Job Titles by Median Salary')
plt.xlabel('Median Salary (USD)')
plt.ylabel('Job Title')

plt.show("\n\n")

# Salary vs Experience Level
plt.figure(figsize=(6,4))
sns.boxplot(x='experience_level', y='salary_median', data=df, palette='coolwarm')
plt.title('Salary vs Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show("\n\n")

# -----------------------------------------------------
# 📅 Step 7. Trends Over Time
# -----------------------------------------------------
# Number of job postings by month
monthly_jobs = df.groupby(df['posted_date'].dt.to_period('M')).size()
plt.figure(figsize=(10,5))
monthly_jobs.plot(kind='line', marker='o', color='coral')
plt.title('Job Postings Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Postings')
plt.grid(True)
plt.show("\n\n")

# Hiring trends for selected roles (safely)
trend = df.groupby([df['posted_date'].dt.to_period('M'), 'job_title']).size().unstack().fillna(0)
top_roles = ['Data Scientist', 'Machine Learning Engineer', 'Data Engineer']
existing_roles = [role for role in top_roles if role in trend.columns]

if existing_roles:
    trend[existing_roles].plot(figsize=(10,6))
    plt.title('Hiring Trends Over Time')
    plt.xlabel('Month')
    plt.ylabel('Number of Postings')
    plt.legend(title='Job Title')
    plt.show("\n\n")
else:
    print("None of the selected top roles exist in the dataset.")

# -----------------------------------------------------
# 🧠 Step 8. Skill Analysis
# ----------------------------------------------------
# Detect skill column
skills_col = None
for col in ['skills_required', 'tools_preferred']:
    if col in df.columns:
        skills_col = col
        break

if skills_col:
    all_skills = []
    for s in df[skills_col].dropna():
        skills = re.split(r',|;|\|', s)
        all_skills.extend([skill.strip().lower() for skill in skills if skill.strip()])

    skills_count = pd.Series(all_skills).value_counts().head(15)

    plt.figure(figsize=(10,5))
    sns.barplot(x=skills_count.values, y=skills_count.index, palette='plasma')
    plt.title('Top 15 In-Demand Skills')
    plt.xlabel('Frequency')
    plt.ylabel('Skill')
    plt.show("\n\n")

    # Skills vs Median Salary
    skill_salary = []
    for s, sal in zip(df[skills_col].dropna(), df['salary_median'].dropna()):
        skills = re.split(r',|;|\|', s)
        for skill in skills:
            skill_salary.append([skill.strip().lower(), sal])
    skill_salary_df = pd.DataFrame(skill_salary, columns=['skill', 'salary'])
    top_skill_salary = skill_salary_df.groupby('skill')['salary'].median().sort_values(ascending=False).head(15)

    plt.figure(figsize=(10,5))
    sns.barplot(x=top_skill_salary.values, y=top_skill_salary.index, palette='crest')
    plt.title('Top 15 Skills by Median Salary')
    plt.xlabel('Median Salary (USD)')
    plt.ylabel('Skill')
    plt.show("\n\n")

# WordCloud of job titles
text = " ".join(df['job_title'].dropna())
plt.figure(figsize=(10,5))
plt.imshow(WordCloud(background_color='white', colormap='viridis').generate(text))
plt.axis('off')
plt.title('Most Common Words in Job Titles')
plt.show("\n\n")

# -----------------------------------------------------
# 🧾 Step 9. Summary Insights
# -----------------------------------------------------
print("\n===== SUMMARY =====")
print("Total Job Listings:", len(df))
print("Date Range:", df['posted_date'].min(), "to", df['posted_date'].max())
print("\nTop 5 Job Titles:\n", df['job_title'].value_counts().head())
print("\nTop 5 Locations:\n", df['location'].value_counts().head())
print("\nOverall Median Salary (USD):", round(df['salary_median'].median(), 2))
print("\nTop 10 Skills:\n", skills_count.head(10) if skills_col else "No skills column found")
print("\n✅ Analysis Complete!")

# -----------------------------------------------------
# 🔬 Step 10. Spark & Machine Learning Example
# -----------------------------------------------------
# Spark DataFrame
spark = SparkSession.builder.appName("AI Job Market").getOrCreate()
spark_df = spark.read.csv('/content/drive/MyDrive/Python/ai_job_market.csv', header=True, inferSchema=True)
spark_df.show(5)

# Simple Linear Regression: Salary prediction based on job title and experience
df_model = df[['salary_median', 'job_title', 'experience_level']].dropna()
le1, le2 = LabelEncoder(), LabelEncoder()
df_model['job_title'] = le1.fit_transform(df_model['job_title'])
df_model['experience_level'] = le2.fit_transform(df_model['experience_level'])

X = df_model[['job_title', 'experience_level']]
y = df_model['salary_median']

model = LinearRegression()
model.fit(X, y)
print("Model Score (R²):", round(model.score(X, y), 3))

# Interactive plot using Plotly
fig = px.scatter(df, x='experience_level', y='salary_median', color='job_title')
fig.show()

# =====================================================
# 🏁 End of Notebook
# =====================================================
```
